Import Necessary Libraries

```python

import pandas as pd
import os
```

Merge .xlsx files into one .csv file

```python

files = [file for file in os.listdir("./Accounts")]
all_accounts = pd.DataFrame()
for file in files:
    df = pd.read_excel("./Accounts/"+file)
    all_accounts = pd.concat([all_accounts, df], sort = True)
all_accounts.to_csv("all_data.csv", index = False)

```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

Making a usufol header

```python

all_data = all_data.drop([0, 1], axis=0)
new_header = all_data.iloc[0]
all_data = all_data[1:]
all_data.columns = new_header
all_data.head()
```

```python
all_data.info()
```

Separation of necessary columns

```python

all_data = all_data[["PrimalacMesto","UkupnaCena"]]
all_data.head()
```

```python
all_data.info()
```

Remove of NaN values for casting column "UkupnaCena" to numeric

```python

nan_df = all_data[all_data.isna().any(axis=1)]
nan_df
all_data = all_data.dropna(axis=0, how="any")
all_data.head()
```

```python
all_data.info()
```

Remove of non-numeric type of values from column "UkupnaCena" for casting to numeric

```python

all_data = all_data[all_data["UkupnaCena"] != "UkupnaCena"]
all_data.head()
```

```python
all_data.info()
```

Cast column "UkupnaCena" to numeric

```python

all_data["UkupnaCena"] = pd.to_numeric(all_data["UkupnaCena"])
all_data.info()
```

Checking of required data

```python
all_data.groupby("PrimalacMesto").sum()*0.80
```

```python
all_data.head()
```

Droping of cells with "BOR BO" and "NEGOTIN BO" which are errors in AKS invoice

```python
all_data = all_data[(all_data.PrimalacMesto == "BEOGRAD BG") | (all_data.PrimalacMesto == "CACAK 1 CA") | \
                   (all_data.PrimalacMesto == "CACAK CA") | (all_data.PrimalacMesto == "KRAGUJEVAC") | \
                   (all_data.PrimalacMesto == "NIS NI") | (all_data.PrimalacMesto == "NISKA BANJA NI") | \
                   (all_data.PrimalacMesto == "NOVI PAZAR NP") | (all_data.PrimalacMesto == "NISKA BANJA NI") | \
                   (all_data.PrimalacMesto == "NOVI SAD NS") | (all_data.PrimalacMesto == "VETERNIK NS") | \
                   (all_data.PrimalacMesto == "ZEMUN ZE")]
all_data.head()
```

```python
all_data.groupby("PrimalacMesto").sum()*0.80
```

```python
all_data.head()
```

Rename of the rows to have a clear name of Cites when we sum the values

```python
all_data["PrimalacMesto"] = all_data["PrimalacMesto"].replace({"ZEMUN ZE": "BEOGRAD"})
all_data["PrimalacMesto"] = all_data["PrimalacMesto"].replace({"BEOGRAD BG": "BEOGRAD"})
all_data["PrimalacMesto"] = all_data["PrimalacMesto"].replace({"VETERNIK NS": "NOVI SAD"})
all_data["PrimalacMesto"] = all_data["PrimalacMesto"].replace({"NOVI SAD NS": "NOVI SAD"})
all_data["PrimalacMesto"] = all_data["PrimalacMesto"].replace({"CACAK 1 CA": "CACAK"})
all_data["PrimalacMesto"] = all_data["PrimalacMesto"].replace({"CACAK CA": "CACAK"})
all_data["PrimalacMesto"] = all_data["PrimalacMesto"].replace({"NISKA BANJA NI": "NIS"})
all_data["PrimalacMesto"] = all_data["PrimalacMesto"].replace({"NIS NI": "NIS"})
all_data["PrimalacMesto"] = all_data["PrimalacMesto"].replace({"NOVI PAZAR NP": "NOVI PAZAR"})
all_data.head()
```

The result for all the Cites, how much they spend on "AKS EXPRESS" delivery

```python
result = all_data.groupby("PrimalacMesto").sum()*0.80
print(result)
```
